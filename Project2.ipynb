```python
#Import
import google.generativeai as genai
import matplotlib.pyplot as plt
import time
```

```python
# Load API Key from Colab Secrets

from google.colab import userdata
api_key = userdata.get('OPENAI_API_KEY')

genai.configure(api_key=api_key)
```

```python
# Find Available Model

models = [m.name for m in genai.list_models() if "generateContent" in m.supported_generation_methods]

print("Available Models:", models)

model_name = models[0]   # pick first available working model
print("Using Model:", model_name)

model = genai.GenerativeModel(model_name)
```

```python
# Topics
topics = [
    "Artificial Intelligence",
    "Machine Learning",
    "Deep Learning",
    "Neural Networks",
    "Natural Language Processing"
]


times = []

for i, topic in enumerate(topics):
    start = time.time()

    response = model.generate_content(
        f"Explain the following concept in simple terms: {topic}"
    )

    end = time.time()

    times.append(end - start)

    print(f"\nTopic {i+1}: {topic}")
    print(f"Response: {response.text.strip()}")
```

```python

# Plot Graph
plt.figure()
plt.plot(range(1, len(times)+1), times, marker='o')

plt.title("Response Time vs Topics (Gemini)")
plt.xlabel("Topics")
plt.ylabel("Time (seconds)")
plt.xticks(range(1, len(topics)+1), ["AI","ML","DL","NN","NLP"])

plt.grid(True)
plt.show()
```
